```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = "20221225.CSV"
df = pd.read_csv(file_path, sep = "\t", index_col = False)
df.head()
```

```python
df["Time"] = pd.to_datetime(df['Time']).dt.time
df.set_index("Time", inplace = True)

df.head()
```

```python
fig, axs = plt.subplots(nrows = 3)
axs = axs.flat

sns.lineplot(x = df.index.astype(str), y = df["Temperature (C)"], ax = axs[0])
axs[0].get_xaxis().set_visible(False)

sns.lineplot(x = df.index.astype(str), y = df["Humidity (%)"], ax = axs[1])
axs[1].get_xaxis().set_visible(False)

sns.lineplot(x = df.index.astype(str), y = df["CO2 (ppm)"], ax = axs[2])
axs[2].xaxis.set_major_locator(plt.MaxNLocator(6))
```
